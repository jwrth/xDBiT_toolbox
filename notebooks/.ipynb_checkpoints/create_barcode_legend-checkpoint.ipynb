```python
import pandas as pd
import numpy as np
```

```python
legend = pd.read_csv("./37_28_barcodes_legend_empty.csv")
assign = pd.read_csv("./37_28_well_coord_assignment.csv")
```

```python
assign.head()
```

```python
coord_names = [n for n in ['X', 'Y', 'Z'] if n in legend.columns]
```

```python
def create_pos_to_coord_dict(assignment_dataframe, coord_name):
    d = {pos: coord for (pos, coord) in zip(assignment_dataframe[coord_name + "_WellPosition"], assignment_dataframe[coord_name + "_Coord"]) if not pd.isnull(pos)}
    return d
```

```python
pos_to_coords = {name: create_pos_to_coord_dict(assign, name) for name in coord_names}
```

```python
def fill_legend(legend, pos_to_coords, name):
    new_column = []
    
    for pos in legend["WellPosition"]:
        if pos in pos_to_coords[name]:
            entry = pos_to_coords[name][pos]
            new_column.append(entry)
        else:
            new_column.append(pd.NA)
    
    legend[name] = new_column
```

```python
for name in coord_names:
    fill_legend(legend, pos_to_coords, name)
```
